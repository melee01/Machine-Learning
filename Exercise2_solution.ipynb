# Problem 1 Solution

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
train_df = pd.read_csv('train.csv')
train_df
```

```python
print('Data types: ',train_df.dtypes)
```

```python
train_df.dtypes
```

```python
numerical_list = []
numerical_df = pd.DataFrame(numerical_list)
```

```python

for collumns in train_df:
    save = train_df[collumns].dtypes
    if(save == 'int64'):
        numerical_df = pd.concat([numerical_df,train_df[collumns]],axis = 1)
```

```python
numerical_df
```

##### Find non-numerical value : 

#### MSSubClass, OverallQual, OverallCond

```python
numerical_df = numerical_df.drop('MSSubClass',axis = 1)
```

```python
numerical_df = numerical_df.drop('OverallQual',axis = 1)
```

```python
numerical_df = numerical_df.drop('OverallCond',axis = 1)
```
