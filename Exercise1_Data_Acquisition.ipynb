# B

```python
top_20_countries = {}
for year in range(2000, 2017, 4):
    url = f'http://en.wikipedia.org/wiki/{year}_Summer_Olympics_medal_table'
    html_data = pd.read_html(url)
    medal_table = html_data[3]
    top_20_countries[year] = medal_table.head(20)
```

```python
for year, top_countries_df in top_20_countries.items():
    print(f"\nTop 20 countries in {year} Summer Olympics:")
    print(top_countries_df)
```
