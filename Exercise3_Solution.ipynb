# Sub-Question A

```python
train_df = pd.read_csv('train.csv')
train_df
```

## Lot Area

```python
LotArea_df = train_df['LotArea']
LotArea_df
```

```python
LotArea_df.describe()
```

```python
# store q1, q2, q3 value
q1 = LotArea_df.describe()['25%']
q2 = LotArea_df.describe()['50%']
q3 = LotArea_df.describe()['75%']
LeftLimit = q1 - 1.5*(q3-q1)
RightLimit = q3 + 1.5 *(q3-q1)

outliers = 0
for data in LotArea_df:
    if data < LeftLimit or data > RightLimit:
        outliers = outliers + 1
outliers
```

```python
f = plt.figure(figsize=(24, 4))
sb.boxplot(data=LotArea_df, orient='h')
```

#### Answer = GrLivArea and GarageArea has the strongest correlation with SalePrice. This is due to the last row in the heatmap and among those 4 variables of LotArea, GrLivArea, TotalBsmtSF, and GarageArea, we can see that the top two variables are GrLivArea with 0.71 and GarageArea with 0.62

```python
train_df = pd.read_csv('train.csv')
train_df
```

```python
print("Number of Levels for GarageType", len(train_df['GarageType'].unique()))
print(train_df['GarageType'].value_counts())
```

```python
train_df.loc[train_df['GarageType'].isnull()] = 0
train_df.loc[train_df['GarageType']=='2Types'] = 1
train_df.loc[train_df['GarageType']=='Attchd'] = 1
train_df.loc[train_df['GarageType']=='Basment'] = 1
train_df.loc[train_df['GarageType']=='BuiltIn'] = 1
train_df.loc[train_df['GarageType']=='CarPort'] = 1
train_df.loc[train_df['GarageType']=='Detchd'] = 1
train_df['GarageType']
train_df['GarageType'].info()
print("Number of Levels for GarageType", len(train_df['GarageType'].unique()))
print(train_df['GarageType'].value_counts())
```

```python
gar_exist = train_df['GarageType']
```

```python
jointDF = pd.concat([SalePrice,gar_exist],axis=1).reindex(SalePrice.index)
jointDF
```

```python
print("Num of Levels for GarageType", len(gar_exist.unique()))
print(gar_exist.value_counts())
```

```python
# f = plt.figure(figsize=(6, 24))
sb.boxplot(x='GarageType',y='SalePrice',data=jointDF)
```

#### The boxplot makes it abundantly evident that a garage has an impact on the house's sale price. It demonstrates the strong link between GarageType and SalePrice because there aren't many overlaps. To be clear, a house with GarageType '0' has no garage, while a house with any type of garage has GarageType '1'. As a result, yes, it is evident that homes without garages typically have lower SalePrices than those that do.
